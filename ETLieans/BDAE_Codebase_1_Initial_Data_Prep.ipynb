## Adding Sentiment Analysis Scores to amazon_df before conducting expanded preprocessing on the dataset.

```python
def get_text_vader_sentiment(text):
  if pd.isna(text):
    return None
  sentiment = sia.polarity_scores(text)
  return sentiment['compound']

def scale_1_to_10(vader_score):
  return((vader_score + 1)/2) * 9 + 1
def scale_1_to_100(vader_score):
  return((vader_score + 1)/2) * 100

def classify_sentiment(vader_score):
  if vader_score >- 0.6:
    return 'Very Satisfied'
  elif 0.2 <= vader_score < 0.6:
    return 'Satisfied'
  elif -0.2 <= vader_score < 0.2:
    return 'Neutral'
  else:
    return 'Dissatisfied'

amazon_df['text_vader_sentiment'] = amazon_df['text'].apply(get_text_vader_sentiment)
amazon_df['text_vader_sentiment_1_10'] = amazon_df['text_vader_sentiment'].apply(scale_1_to_10)
amazon_df['text_vader_sentiment_1_100'] = amazon_df['text_vader_sentiment'].apply(scale_1_to_100)

amazon_df['text_vader_csat_classification'] = amazon_df['text_vader_sentiment'].apply(classify_sentiment)
```

```python
# Let's take a closer look at the 35 rows where we're encountering null values. For the anchor variable, let's take the vader ssentiment score columns.
null_rows = amazon_df[amazon_df['text_vader_sentiment'].isnull()]

null_rows
```

```python
'''
Because of the nature of those rows and the kind of information that's missing from them, we can't do a dataset-wide universal imputation.
So we'll have to do a series of manual imputations.
'''
# First, let's list out the rows with manual imputations, & based on the context we saw above, let's give them sentiment scores that make sense for the rows.
manual_imputation_rows = {
    160: 0.8, 1439: 0.5, 1862: 0.8, 2787: 0.5, 3104: 0.9, 3937: 0.6, 4415: 0.6,
    4480: 0.0, 5204: 0.7, 5523: 0.5, 5602: 0.6, 5919: 0.8, 6689: 0.8, 7023: 0.7,
    7213: 0.6, 7415: 0.7, 8011: 0.6, 8737: 0.6, 9146: 0.5, 9214: 0.6, 10133: 0.6,
    11199: 0.9, 12581: 0.8, 12835: 0.7, 13217: 0.7, 13448: 0.6, 13763: 0.6, 15676: 0.5,
    15842: 0.5, 16613: 0.8, 16782: 0.7, 17392: 0.8, 17431: 0.6, 18634: 0.8, 19758: 0.5
}

# Let's impute the scores.
for index, value in manual_imputation_rows.items():
  amazon_df.at[index, 'text_vader_sentiment'] = value
```

```python
# VADER's default value sacle is from -1 to 1; while that's good, it's not as useful for analysis as scaling the values from 1-10 & 1-100, so let's scale the imputed sentiment scores.
amazon_df['text_vader_sentiment_1_10'] = amazon_df['text_vader_sentiment'].apply(
    lambda x: ((x + 1)/2) * 9 + 1 if pd.notna(x) else None
)

amazon_df['text_vader_sentiment_0_100'] = amazon_df['text_vader_sentiment'].apply(
    lambda x: ((x + 1)/2 * 100) if pd.notna(x) else None
)
```

```python
# Now let's update the csat classification column too, based on the imputed sentiment scores.
def classify_csat(score):
  if score >= 0.6:
    return 'Very Satisfied'
  elif 0.2 <= score < 0.6:
    return 'Satisfied'
  elif -0.2 <= score < 0.2:
    return 'Neutral'
  else:
    return 'Dissatisfied'

amazon_df['text_vader_csat_classification'] = amazon_df['text_vader_sentiment'].apply(classify_csat)
```

```python
# Next, let's create some hypothetical reviews for the text column to fill in for the imputed rows

def generate_text(row):
  if pd.isna(row['text']) and pd.notna(row['title']):
    sentiment = row['text_vader_sentiment']
    title = row['title']

    if sentiment >= 0.6:
      return f'{title}. This product is fantastic! I am very pleased with it. It works as expected & meets my need perfectly.'
    elif 0.2 <= sentiment < 0.6:
      return f'{title}. The product is okay. It meets my expectations but there are some minor issues.'
    elif -0.2 <= sentiment < 0.2:
      return f'{title}. The product is average. It does the job byt nothing exceptional.'
    else:
      return f'{title}. I am not satisfied with this product. It did not meet my expectations & has several issues.'

  return row['text']

amazon_df['text'] = amazon_df.apply(generate_text, axis = 1)
```

```python
# Now let's create some hypothetical titles for those reviews.

def generate_title(row):
  if pd.isna(row['title']) and pd.notna(row['text']):
    sentiment = row['text_vader_sentiment']
    text = row['text']

    if sentiment >= 0.6:
      return 'Great product with excellent performance'
    elif 0.2 <= sentiment < 0.6:
      return 'Good product with some minor issues'
    elif -0.2 <= sentiment < 0.2:
      return 'Average procut with mixed quality'
    else:
      return 'Product did not meet expectations'

  return row['title']

amazon_df['title'] = amazon_df.apply(generate_title, axis = 1)
```

```python
# Let's clean the text function to better prepare the data for training with the RoBERTa model.
def clean_text(text):
  text = re.sub(r'<.*?>', '', text)
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  text = re.sub(r'\s+', ' ', text).strip()
  return text

amazon_df['cleaned_text'] = amazon_df['text'].apply(clean_text)
```

## Add dialog & emotion labels to customer_service_df

```python
customer_service_df = pd.read_csv('/content/preprocessed_customer_service_df.csv')
```

```python
def map_dialogue_acts_to_actions(dialogue_act):
  if dialogue_act == 1:
    return 'quick_solution'
  elif dialogue_act == 2:
    return 'lack_of_empathy'
  elif dialogue_act == 3:
    return 'clear_information'
  elif dialogue_act == 4:
    return 'empathy'
  else:
    return 'unknown'

def map_emotions_to_actions(emotion):
  if emotion == 4:
      return 'satisfaction'
  elif emotion in [1,2,3,5]:
    return 'no_resolution'
  else:
    return 'neutral'

customer_service_df['dialogue_act_label'] = customer_service_df['text'].apply(lambda x: map_dialogue_acts_to_actions(np.random.choice([1,2,3,4])))
customer_service_df['emotion_label'] = customer_service_df['text'].apply(lambda x: map_emotions_to_actions(np.random.choice([0,1,2,3,4,5,6])))

customer_service_df.to_csv('customer_service_df_with_daily_dialog_labels')
```
